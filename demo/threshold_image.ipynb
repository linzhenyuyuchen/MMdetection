```python
import glob
import numpy as np
import cv2
import matplotlib.pylab as plt
```

```python
cocoRoot= "/data2/lzy/pathology/train2017/"
ls = glob.glob(cocoRoot+"*")
```

```python
for i in ls:
    img = cv2.imread(i)
    GrayImage = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret,thresh1 = cv2.threshold(GrayImage,123,255,cv2.THRESH_BINARY)
    img = cv2.medianBlur(thresh1,5)
    cv2.imwrite(i.replace("train2017/","train2017_2/"),img)
```
